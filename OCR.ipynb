```python
# import necessary libraries
import cv2  # for image reading, filtering
from pytesseract import pytesseract # for OCR
from matplotlib import pyplot as plt # for displaying image
```

#High Quality Image

##Traditional OCR

```python
# path for image
image_path = "/content/high_quality.jpg"

# extracting text from the image using image_to_string() function
text = pytesseract.image_to_string(image_path)
  
# print extracted text
print(text)
```

#Poor Quality Image

##Traditional OCR

```python
custom_config = r'-l eng -psm 6' 

# path for image from google drive
image_path = "/content/poor_quality.jpg"

# extracting text from the image using image_to_string() function
text = pytesseract.image_to_string(image_path, config=custom_config)
  
# print extracted text
print(text)
```

##Filtering

Gaussian Filter

```python
# read image using cv2
img = cv2.imread('/content/poor_quality.jpg')

# apply gaussian filter
G_filt_img = cv2.GaussianBlur(img,(1,1),0)

# Extract text using OCR
text_G = pytesseract.image_to_string(G_filt_img, config=custom_config)

print(text_G)
```

Bilateral Filter

```python
# read image using cv2
img = cv2.imread('/content/poor_quality.jpg')

# apply Bilateral filter
B_filt_img = cv2.bilateralFilter(img,5,50,50)

# Extract text using OCR
text_B = pytesseract.image_to_string(B_filt_img, config=custom_config)

print(text_B) 
```
